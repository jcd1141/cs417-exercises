```python
from math import gcd
class Fraction:
  def __init__(self,top,bottom):
    self.num = top
    self.den = bottom
  def __str__(self):
    return f"{self.num}/{self.den}"
  def _reduce(self):
    common_d = gcd(abs(self.num), abs(self.den))
    self.num //= common_d
    self.den //= common_d
    return self
  def __add__(self,other):
    new_num = self.num*other.den + self.den * other.num
    new_den = self.den * other.den
    result = Fraction(new_num,new_den)
    return result._reduce()
  def __sub__(self,other):
    new_num = self.num*other.den - self.den * other.num
    new_den = self.den * other.den
    result = Fraction(new_num,new_den)
    return result._reduce()
  def __mul__(self,other):
    new_num = self.num*other.num
    new_den = self.den*other.den
    result = Fraction(new_num,new_den)
    return result._reduce()
  def __truediv__(self,other):
    new_num = self.num * other.den
    new_den = self.den * other.num
    result = Fraction(new_num,new_den)
    return result._reduce()
  def __eq__(self,other):
    self_reduced = Fraction(self.num,self.den)._reduce()
    other_reduced = Fraction(other.num,other.den)._reduce()
    return (self_reduced.num == other_reduced.num and self_reduced.den == other_reduced.den)

#__add__ demonstration
f5 = Fraction(1,4)
f6 = Fraction(1,2)
result = f5 + f6
print(f"{f5} + {f6} = {result}")

#__sub__ demonstration
f7 = Fraction(3,8)
f8 = Fraction(1,5)
result2 = f7 - f8
print(f"{f7} - {f8} = {result2}")

#__mul__ demonstration
f9 = Fraction(4,5)
f10 = Fraction(2,3)
result3 = f9 * f10
print(f"{f9} * {f10} = {result3}")

#__truediv__ demonstration
f11 = Fraction(3,8)
f12 = Fraction(1,5)
result2 = f11 / f12
print(f"{f11} / {f12} = {result2}")

#equality demonstration
f13 = Fraction(2,4)
f14 = Fraction(1,2)
print(f"{f13} == {f14} = {f13 == f14}")
```
